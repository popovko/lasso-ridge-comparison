```python
from sklearn.preprocessing import scale
from sklearn.utils import shuffle
```

```python
df_shuffled = shuffle(df, random_state = 123)
X = scale(df_shuffled[df_shuffled.columns[:-1]])
y = df_shuffled["cnt"]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_regressor = LinearRegression()
lin_regressor.fit(X, y)
lin_regressor.coef_
for col, coef in zip(df.columns, lin_regressor.coef_):
    print(col, coef)
```

```python
from sklearn.linear_model import Lasso, Ridge
```

```python
lasso_regressor = Lasso()
lasso_regressor.fit(X, y)
for col, coef in zip(df.columns, lasso_regressor.coef_):
    print(col, coef)
```

```python
ridge_regressor = Ridge()
ridge_regressor.fit(X, y)
for col, coef in zip(df.columns, ridge_regressor.coef_):
    print(col, coef)
```

```python
alphas = np.arange(1, 500, 50)
coefs_lasso = np.zeros((alphas.shape[0], X.shape[1]))
coefs_ridge = np.zeros((alphas.shape[0], X.shape[1]))
```

```python
for i in range(alphas.shape[0]):
    ridge_regressor = Ridge(alpha = alphas[i])
    ridge_regressor.fit(X, y)
    coefs_ridge[i] = ridge_regressor.coef_
    
    lasso_regressor = Lasso(alpha = alphas[i])
    lasso_regressor.fit(X, y)
    coefs_lasso[i] = lasso_regressor.coef_
```

```python
plt.figure(figsize=(8, 5))
for coef, feature in zip(coefs_lasso.T, df.columns):
    plt.plot(alphas, coef, label=feature, color=np.random.rand(3))
plt.legend(loc="upper right", bbox_to_anchor=(1.4, 0.95))
plt.xlabel("alpha")
plt.ylabel("feature weight")
plt.title("Lasso")

plt.figure(figsize=(8, 5))
for coef, feature in zip(coefs_ridge.T, df.columns):
    plt.plot(alphas, coef, label=feature, color=np.random.rand(3))
plt.legend(loc="upper right", bbox_to_anchor=(1.4, 0.95))
plt.xlabel("alpha")
plt.ylabel("feature weight")
plt.title("Ridge")
```

```python
from sklearn.linear_model import LassoCV
```

```python
alphas = np.arange(1, 100, 5)
lassoCV_regressor = LassoCV(alphas = alphas)
lassoCV_regressor.fit(X, y)
print(lassoCV_regressor.alpha_)
for coef, fearure in zip(lassoCV_regressor.coef_, df.columns):
    print(coef, fearure)
```
